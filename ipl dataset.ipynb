```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named ipl_data with 'match_winner' and 'toss_winner' columns
# Count the number of matches won by each team
match_winner_counts = df['match_winner'].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
match_winner_counts.plot(kind='bar', color='skyblue')
plt.title('Matches Won by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Matches Won')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
#This bar plot will show the number of matches won by each team in the dataset.

```

#### Gujarat team wons the highest number of matches 

```python

toss_winner_counts = df['toss_winner'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(toss_winner_counts, labels=toss_winner_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Toss Wins by Teams')
plt.axis('equal') 
plt.show()
```

##### Winning the toss does not have a significant effect on match outcomes
